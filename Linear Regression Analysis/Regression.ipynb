```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#training part
train_set = pd.read_csv("train.csv", index_col = None, header = 0)
test_set = pd.read_csv("test.csv", index_col = None, header = 0)
n = len(train_set)

#formula we will use  Yi = β0 + β1*Xi

# calculating x̄ and ȳ
x_mean = train_set["x"].mean()
y_mean = train_set["y"].mean()

#calculating SSxy using formula SSxy = Σ(Xi*Yi) - n*x̄*ȳ
x = train_set["x"].to_numpy()
y = train_set["y"].to_numpy()
cross_deviation = x * y - (n*x_mean*y_mean)
SSxy = sum(cross_deviation)

#calculating SSxx using formula SSxx = Σ(Xi)^2 - n*(x̄)^2
squared_deviation = x**2 - (n*(x_mean**2))
SSxx = sum(squared_deviation)

#calculating β1 using formula β1 = SSxy/SSxx
beta1 = SSxy / SSxx
#calculating β0 using formula β0 = ȳ - β1*x̄
beta0 = y_mean - (beta1*x_mean)
# We've got our prediction line
prediction_line = beta0 + beta1*x

# Plotting the training set
plt.scatter('x', 'y', data = train_set.head(60), color = "red")
plt.plot(x, prediction_line, color='tab:green')
plt.title("X vs Y (Training set)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Plotting the testing set
plt.scatter('x', 'y', data = test_set.head(60), color = "red")
plt.plot(x, prediction_line, color='tab:green')
plt.title("X vs Y (Testing set)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
def prediction(x):
    return beta0 + beta1*x

#testing part
n = len(test_set)
test_array = test_set.to_numpy()
errors = np.array([data[1] - prediction(data[0]) for data in test_array])

mean_squared_error = sum(errors**2)
print(mean_squared_error/(2*n))
```
